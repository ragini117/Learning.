### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
x1,y1= int(input("Enter 1st co-ordinate")),int(input("Enter 2nd co-ordinate"))
x2,y2= int(input("Enter 1st co-ordinate")),int(input("Enter 2nd co-ordinate"))
d=((x2-x1)**2+(y2-y1)**2)**0.5
print("Euclidean Distance between two coordinate is:",d)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
x=0
n=int(input("Enter the no :"))
for i in range (1,n+1):
    x=x+i**2
print("Sum of Square =",x)
    
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
a=int(input("Enter 1st no :"))
b=int(input("Enter 2nd no :"))
n=int(input("Enter nth term :"))
d=b-a
x=a+(n-1)*d
print(f"The {n}th term is :",x)
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
#Get dimensions of tank 
import math 

H=float(input("Enter the height of tank (in Cm) :"))
L=float(input("Enter the lenght of tank (in Cm) :"))
B=float(input("Enter the breadth of tank (in Cm) :"))
V_tank=L*B*H
#Get dimension of Glass
h=float(input("Enter the height of glass (in Cm):"))
r=float(input("Enter the radius of glass(in Cm) :"))
V_glass=(3.14)*(r**2)*(h)

no=math.floor(V_tank/V_glass)

print("No of Glass:",no)
````
