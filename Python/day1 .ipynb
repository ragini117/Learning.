**Take marks of 5 subjects as input and print the percentage and grade.**

**If % ≥ 90 → Grade A**

**If % ≥ 75 → Grade B**

**If % ≥ 50 → Grade C**

**Else → Fail**

```python
a=int(input("Enter Marks of 1st subject :"))
b=int(input("Enter Marks of 2nd subject :"))
c=int(input("Enter Marks of 3rd subject :"))
d=int(input("Enter Marks of 4rth subject :"))
e=int(input("Enter Marks of 5th subject :"))
t=a+b+c+d+e
per=t/500*100
if per>=90:
    print(per,"%","Grade A")
elif per>=75:
    print(per,"%" ,"Grade B")
elif per>=50:
    print(per,"%","Grade C")
else:
    print("Fail")
        
```
