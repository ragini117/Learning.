```python
import numpy as np    # for numerical computation 
import pandas as pd    # for data manipulation & dataframe handling 
import matplotlib.pyplot as plt  # for plotting graph & for visualization 
 
```

```python
df=pd.read_csv("placement.csv")   #load csv into dataframe 
```

```python
df.head()     # display top 5 row 
```

```python
df.tail(4)    # display last 4 row 
```

```python
# plot  scatterplot to see if data is sort of linear or not 
plt.scatter(df['cgpa'],df['package'])
plt.xlabel("cgpa")
plt.ylabel("Package")
```

```python
# split the data into 2 coln i/p and o/p coln 
X = df.iloc[:, 0:1]   # select 1st coln 
Y = df.iloc[:, -1]     # select last coln
```

```python
print(X.shape)   # display dimension for x(row,coln)
print(Y.shape)    # display dimension for y 
```

```python
# split dataset in train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
# import linear Regression class
from sklearn.linear_model import LinearRegression
```

```python
#create object of this class 
lr=LinearRegression()
```

```python
#train model
lr.fit(X_train,Y_train)
```

```python
#predict o/p
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
# lets plot best fit line on scatterplot graph
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('cgpa')
plt.ylabel('package')
```

```python
# find coefficient for this line y=mx+c
m=lr.coef_
```
